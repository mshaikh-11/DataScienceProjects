# Initial Investigations

```python
#Load the Housing dataset
import pandas as pd 
Housing = pd.read_csv("downloads/London.csv")
Housing
```

```python
#Load the location dataset: 
import pandas as pd 
Location = pd.read_csv("downloads/open_postcode_geo.csv")
Location
```

```python
#Renaming the columns i want to keep in longitude , latitude and postal code which has the same column name 
# and data values type as the housing dataset to make sure that i can use the pandas merge function and argument
# ,So i can merge the Location dataset with the housing dataset, this will create a new dataframe with 
# the postcode that matches for both datasets which helps clean my dataset as it will remove invalid data location
#  with respect to postal code as the location dataset provides the every valid postcode in the uk, verified by 
# having a uk open goverment licence.

Location = Location.rename(columns={"AB1 0AA":"Postal Code","57.101474":"Longitude","-2.242851":"Latitude"})
Location
```

```python
# Merge the Housing dataset with the location dataset based on the same column name "Postal Code" which is the post code
df = pd.merge(Housing,Location,on="Postal Code")
df
#Result in the rows being reduced from 3480 (from the housing dataset) to 3395 rows, rows removed and consequently 
#rows reduced (using the location dataset) there has also 10 columns added (from the location dataset) 
#Therefore we removed 85 observations as a result due to the invalid postcodes.
```

```python
#Now, i want to drop the unwanted columns such as the different version of postcodes such as innner codes and outer codes 
# codes with the exception of postcode Area which gives direction of location in a City.
# i also want to remove column which provide the index via the location dataset which is now arbitrary.

import pandas as pd 
df = df.drop(columns=["Unnamed: 0","terminated","small","385386","801193","1","Scotland","AB10AA","AB1 0AA.1","AB1 0","AB1.1","0AA","AB1","AB1  0AA"])
df
```
