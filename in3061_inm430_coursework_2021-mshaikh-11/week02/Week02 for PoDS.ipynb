```python
# (Part 2 step 1) Load the slightly modified Titanic survival data into a pandas data frame.
import pandas as pd 
df=pd.read_csv('titanicSurvival_m.csv')
df
```

```python
#(Part 2 step 2) Find the counts of missing values in each column
df.isnull()
df.isnull().sum()
```

```python
#(Part 2 step 3) Compute the mean and other descriptive statistics and note these down
df.describe()
```

```python
#(Part 2 step 4) Replace the missing values in "Age" and "Fare" columns with 0 values, and visualise in a scatterplot
age_null= df["Age"].fillna(0)
Fare_null= df["Fare"].fillna(0)
import matplotlib.pyplot as plt 
import numpy as np
plt.xlabel("Age")
plt.ylabel("Fare")
plt.scatter(age_null,Fare_null)
plt.show()
```

```python
#(Part 2 step 5) Replace the missing values in "Age" and "Fare" columns with the mean of each column, and visualise in a scatterplot
Age_mean=df["Age"].fillna(29.699118) # mean values from previus step in use of the describe function
Fare_mean=df["Fare"].fillna(32.304319)
import matplotlib.pyplot as plt 
import numpy as np
plt.xlabel("Age")
plt.ylabel("Fare")
plt.scatter(Age_mean,Fare_mean)
plt.show()

#(Part 2 step 6) Reflect on the differences you see in these plots
#There is no difference between the scatter plots 

```

# Part 3 Data Transformations

```python
#(Part 3 step 1) Download the csv data file from WHO on Tuberculosis (from Week01). Information on the data can be found on WHO's web page.
import pandas as pd 
f = pd.read_csv('TB_burden_countries_2014-09-29.csv')
f
```

```python
#(Part 3 step 2) You may need to replace missing values before you start.
f.isnull()
f.isnull().sum()

f=f.fillna(f.mean())
```

```python
#(Part 3 step 3) Choose a number of columns with different shapes, for instance, "e_prev_100k_hi" is right skewed and visualise on an histogram
#_prev_100k	e_prev_100k_lo	e_prev_100k_hi	
import matplotlib.pyplot as plt 
plt.hist(f.e_prev_100k_hi)
plt.hist(f.e_prev_100k)
plt.hist(f.e_prev_100k_lo)
```

```python
#(Part 3 step 4) Apply a log transformation on the data. Numpy has a log function. and visualise. Observe the changes
import numpy as np 
np.log(f.e_prev_100k_hi)
np.log(f.e_prev_100k)
np.log(f.e_prev_100k_lo)

plt.hist(np.log(f.e_prev_100k_hi))
plt.hist(np.log(f.e_prev_100k))
plt.hist(np.log(f.e_prev_100k_lo))


```

```python
#(Part 3 step 5) Choose the numerical columns and map all the columns to [0,1] interval
result = f.select_dtypes(include='number')
result
```

```python
#(Part 3 step 5) Choose the numerical columns and map all the columns to [0,1] interval
normalised  = (result - result.min())/(result.max() -  result.min())
normalised
```

```python
#(Part 3 step 6) Now you can compare the means of each column.
a=f.mean()
print("means of dataset original:",a)
b=normalised.mean()
print("means of normalised dataset:",b)
```
